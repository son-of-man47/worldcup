AVERAGE AGE OF EACH TEAM 

```python
age = df.groupby("Team")["Age"].mean().sort_values().reset_index()
age.columns = ["Teams", "Avg_age"]
age["Avg_age"] = age["Avg_age"].round().astype(int)
age.head()
```

```python
sns.barplot(data = age,  x = "Teams", y = "Avg_age")
plt.ylabel("Average Age")
plt.xticks(rotation = 75)
```

PLAYERS WITH OVER 100 CAPS IN EACH COUNTRY

```python
def caps_count(team):
    count = df[(df.Team) == team]
    caps = count[(count.Caps) >= 100]
    return caps
```

```python
caps_count("Spain")
```

HIGHEST GOAL SCORER IN EACH COUNTRY

```python
def highest_scorer(team):
    count = df[(df.Team) == team]
    hgs = count["Goals"].max()
    return count.loc[count["Goals"] == hgs]
```

```python
highest_scorer("Spain")
```

HIGHEST GOAL SCORER IN A TEAM BASED ON POSITIONS (FW, MF, DF)

```python
def highest_scorer_cat(team, pos):
    count = df[(df.Team) == team]
    pos = count[(count.Position) == pos]
    hgs = pos["Goals"].max()
    return count.loc[count["Goals"] == hgs]
```

```python
highest_scorer_cat("Spain", "FW")
```
